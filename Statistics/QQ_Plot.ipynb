```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
# how looks distribution
sns.kdeplot(df['sepal_length'])
```

```python
# sort the sepal length value and save it
temp = sorted(df['sepal_length'].tolist())
```

```python
# calculate Sample Quantiles
y_quant = []
for i in range(1,101):
  y_quant.append(np.percentile(temp,i))
```

```python
# generate normal distributed data (mean=0, std=1)
samples = np.random.normal(loc=0,scale=1,size=1000)
```

```python
# calculate Therotical Quantiles
x_quant = []
for i in range(1,101):
  x_quant.append(np.percentile(samples, i))
```

```python
# plot scatter plot
sns.scatterplot(x = x_quant,y = y_quant)
```

```python
# using statmodel (another method)
import statsmodels.api as sm
import matplotlib.pyplot as plt
# create a QQ plot of two sets of data
fig = sm.qqplot(df['sepal_length'],line='s',fit=True)
plt.title('QQ Plot')
plt.xlabel('Therotical Quantiles')
plt.ylabel('Sample Quantiles')
plt.show()
```

```python
# generate 1000 set of random data
x = np.random.uniform(low=0,high=1,size=1000)
plt.hist(x)
plt.show()
```

```python
from scipy.stats import uniform
# fit a uniform distribution to the data

params = uniform.fit(x)
params
```

```python
distribution = uniform(loc=params[0],scale=params[1])
```

```python
# create QQ plot of data using the uniform distribution
fig = sm.qqplot(x, dist=distribution,line='45')
plt.title('QQ Plot')
plt.xlabel('Therotical Quantiles')
plt.ylabel('Sample Quantiles')
plt.show()
```
