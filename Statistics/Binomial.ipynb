```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# number of trials
n=10
# probability of success
p=0.5
# number of samples generate
size=1000
# make binomial distribution
binomial_distribution = np.random.binomial(n, p, size)
# plot binomial distribution
plt.hist(binomial_distribution, density=True)
plt.show()
```

```python
# number of trials
trials=10
# probability of success
probabilities=[0.5,0.8,0.2]
# number of samples generate
size=1000

a = 1  # number of rows
b = 3  # number of columns
c = 1  # initialize plot counter

# figure size
fig = plt.figure(figsize = (15,3))
plt.suptitle("Binomial Distribution", fontsize = 18)

for probability in probabilities:
    plt.subplot(a, b, c)
    # xlabel
    plt.xlabel('probability of {}'.format(probability))
    # make binomial distribution
    binomial_distribution = np.random.binomial(trials, probability, size)
    # plot binomial distribution
    plt.hist(binomial_distribution, density=True)
    c = c + 1

plt.tight_layout()
plt.show()
```
