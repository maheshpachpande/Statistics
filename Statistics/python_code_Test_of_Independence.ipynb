```python
import numpy as np
import pandas as pd
from scipy.stats import stats
```

```python
# We will use chi-square Test-square-Independence to see survival rate of Passengers is independence on Pclass
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
df.head()
```

```python
# Null Hypothesis (Ho) : There is no significant association between Passengers is independence on Pclass
# Alternate Hypothesis (H1) : There is significant association between Passengers is independence on Pclass
```

```python
# create contingency table
contingency_table = pd.crosstab(df['Survived'],df['Pclass'])
contingency_table
```

```python
from scipy.stats.contingency import chi2_contingency
# chi-square Test-square-Independence to see survival rate of Passengers is independence on Pclass
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

print('chi2 statistics : {:.2f}'.format(chi2))
print('p_value :{:.2f}'.format(p_value))
print('dof :{}'.format(dof))
print('expected : \n{}'.format(expected))
```

```python
# Interpret the result
alpha = 0.05
if p_value < alpha:
  print('Reject the Null Hypothesis')
else:
  print('Accept the Null Hypothesis')

print('='*35)
```
