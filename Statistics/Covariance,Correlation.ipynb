Covariance defines the directional association between the variables.

Covariance values range from -inf to +inf

a negative value denotes that both the variables move in opposite directions."""

Correlation is a statistical measure that expresses the extent to which two variables are linearly related (meaning how much they change together at a constant rate).
The strength and directional association of the relationship between two variables are defined by correlation and it ranges from -1 to +1.
Similar to covariance, a positive value denotes that both variables move in the same direction whereas a negative value tells us that they move in opposite directions.

The correlation coefficient is a statistical measure of the strength of a linear relationship between two variables.

Its values can range from -1 to 1.

A correlation coefficient of -1 describes a perfect negative,
 or
  
A correlation coefficient of 1 describes a perfect positive.
