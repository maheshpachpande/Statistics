```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# rolling a dice 10000 times and getting 10000 output between 1 to 6
l=[]
for i in range(10000):
  l.append(random.randint(1,6))
```

```python
len(l)
```

```python
# first 5 times rolling a dice and getting output
l[:5]
```

```python
# probability of each outcome
s = pd.Series(l)
s = (s.value_counts()/s.value_counts().sum()).sort_index()
```

```python
# plot graph
s.plot(kind='bar',title='PMF of rolling Dice with probability')
```

```python
np.cumsum(s)
```

```python
plt.figure(figsize=(10,3))
plt.subplot(121)
s.plot(kind='bar',title='PMF of rolling Dice with probability')
plt.subplot(122)
np.cumsum(s).plot(title='CDF of single rolling Dice with probability density')
plt.show()
```

```python
# Rolling Two different Dice together 10000 times
m = []
for i in range(10000):
  a = random.randint(1,6)
  b = random.randint(1,6)
  m.append(a+b)
```

```python
# first 5 times roll dice and getting output
m[:5]
```

```python
# probability of each outcome
d = pd.Series(m)
d = (d.value_counts()/d.value_counts().sum()).sort_index()
```

```python
# plot graph for two dice
d.plot(kind='bar',title='PMF for Two Rolling Dice with probability')
```

```python
# cumulative distribution function (cdf of pmf)
np.cumsum(d)
```

```python
np.cumsum(d).plot()
```

```python
plt.figure(figsize=(10,3))
plt.subplot(121)
d.plot(kind='bar',title='PMF for Two Rolling Dice with probability')
plt.subplot(122)
np.cumsum(d).plot(title='cdf for two rolling dice with probability density')
plt.show()
```
