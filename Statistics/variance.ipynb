```python
import statistics as st
import seaborn as sns
import numpy as np
```

```python
import matplotlib.pyplot as plt
# first data
data_1=np.random.normal(size=1000, scale=1.0, loc=0.0)
# variance of first data
variance = st.variance(data_1)
print('Variance data_1 : ',variance)
# figure
plt.figure(figsize=(5,2))
# plot first data
sns.kdeplot(data=data_1, label='data_1')
# second data
data_2=np.random.normal(size=1000, scale=3.0, loc=0.0)
# variance of second data
variance_2 = st.variance(data_2)
print('Variance data_2 : ',variance_2)
# plot second data
sns.kdeplot(data=data_2,label='data_2')
# give titile
plt.title('Showing Variance')
# add label info
plt.legend(loc='upper right')
# add a grid...
plt.grid(True)
# show the plot....
plt.show()
```

```python
# calculate standard deviation of first data
standard_deviation_1 = st.stdev(data_1)
print('Standard Deviation of data_1 : ',standard_deviation_1)
print('='*35) # separator
# calculate mean of first data
mean_1 = np.mean(data_1)
print('Mean_1 : ',mean_1)
print('='*35)
# first standard deviation above and below mean
first_std_above = mean_1 + standard_deviation_1
first_std_below = mean_1 - standard_deviation_1
# sec standard deviation above and below mean
sec_std_above = mean_1 + (standard_deviation_1*2)
sec_std_below = mean_1 - (standard_deviation_1*2)
# third standard deviation above and below mean
third_std_above = mean_1 + (standard_deviation_1*3)
third_std_below = mean_1 - (standard_deviation_1*3)

print('first std above mean of data_1 :',first_std_above)
print('first std below mean of data_1 :',first_std_below)
print('='*35)
print('sec std above mean of data_1 :',sec_std_above)
print('sec std below mean of data_1 :',sec_std_below)
print('='*35)
print('third std above mean of data_1 :',third_std_above)
print('third std below mean of data_1 :',third_std_below)
print('='*35)
```

```python
# calculate standard deviation of second data
standard_deviation_2 = st.stdev(data_2)
print('Standard Deviation of data_2 : ',standard_deviation_2)
print('='*35)
# calculate mean of second data
mean_2 = np.mean(data_2)
print('Mean_2 : ',mean_2)
print('='*35)
# first standard deviation above and below mean
f_std_above = mean_2 + standard_deviation_2
f_std_below = mean_2 - standard_deviation_2
# second standard deviation above and below mean
s_std_above = mean_2 + (standard_deviation_2*2)
s_std_below = mean_2 - (standard_deviation_2*2)
# third standard deviation above and below mean
t_std_above = mean_2 + (standard_deviation_2*3)
t_std_below = mean_2 - (standard_deviation_2*3)

print('first std above mean of data_1 :',f_std_above)
print('first std below mean of data_1 :',f_std_below)
print('='*35)
print('sec std above mean of data_1 :',s_std_above)
print('sec std below mean of data_1 :',s_std_below)
print('='*35)
print('third std above mean of data_1 :',t_std_above)
print('third std below mean of data_1 :',t_std_below)
print('='*35)
```

```python
plt.figure(figsize=(10,2.5))
# plot first data
sns.histplot(x=data_1,color='lightgreen',edgecolor='lightgreen')
# plot mean vertical line
plt.axvline(x=mean_1, ymin=0, ymax=max(data_1),color='black',label= 'x̄')
# plot vertical line of first standard deviation
plt.axvline(x=first_std_below,linestyle='dashed',label='I = x̄-σ',color='red')
plt.axvline(x=first_std_above,linestyle='dashed',label='I = x̄+σ',color='red')
# plot vertical line of second standard deviation
plt.axvline(x=sec_std_below,linestyle='dashed',label='II = x̄-σ',color='violet')
plt.axvline(x=sec_std_above, linestyle='dashed',label='II = x̄+σ',color='violet')
# plot vertical line of third standard deviation
plt.axvline(x=third_std_below, linestyle='dashed',label='III = x̄-σ')
plt.axvline(x=third_std_above, linestyle='dashed',label='III = x̄+σ')

# add legend...
plt.legend(bbox_to_anchor = (1.0, 1),loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(10,2.5))
# plot first data
sns.histplot(x=data_2,color='lightgreen',edgecolor='lightgreen')
# plot mean vertical line
plt.axvline(x=mean_2, ymin=0, ymax=max(data_2),color='black',label= 'x̄')
# plot vertical line of first standard deviation
plt.axvline(x=f_std_below,linestyle='dashed',label='I = x̄-σ',color='red')
plt.axvline(x=f_std_above,linestyle='dashed',label='I = x̄+σ',color='red')
# plot vertical line of second standard deviation
plt.axvline(x=s_std_below,linestyle='dashed',label='II = x̄-σ',color='yellow')
plt.axvline(x=s_std_above, linestyle='dashed',label='II = x̄+σ',color='yellow')
# plot vertical line of third standard deviation
plt.axvline(x=t_std_below, linestyle='dashed',label='III = x̄-σ',color='gray')
plt.axvline(x=t_std_above, linestyle='dashed',label='III = x̄+σ',color='gray')

# add legend...
plt.legend(bbox_to_anchor = (1.0, 1),loc='upper right')
plt.show()
```
