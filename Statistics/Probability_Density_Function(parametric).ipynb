```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generate 1000 datapoints
sample = np.random.normal(loc=50, scale=5, size=1000)
```

```python
sample.mean()
```

```python
sample.std()
```

```python
# plot histogram to understand the distribution of data
plt.figure(figsize=(10,3))
plt.hist(sample,bins=10)
plt.show()
```

```python
# Calculate the sample mean and sample standard deviation
sample_mean = sample.mean()
sample_std = sample.std()
```

```python
# fit the distribution above parameter
distribution = norm(sample_mean,sample_std)
```

```python
# generate 100 datapoints between min value of sample and max value of sample
values = np.linspace(sample.min(),sample.max(),100)
```

```python
# calculate probability density function
probability_density = [distribution.pdf(value) for value in values]
```

```python
# plot the histogram and pdf
plt.figure(figsize=(10,3))
plt.hist(sample,bins=10,density=True)
plt.plot(values,probability_density)
plt.show()
```
